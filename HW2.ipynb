## Задание 1

```python
#первое задание 

text = "варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг" 

#разделяем текст из переменной text на слова 
words = text.split() 
bigramy = [] 

#создаем биграммы 

#проходимся по всем словам, не считая последнего 
for word in range(len(words)-1): 
    bigramy.append(words[word] + " " + words[word+1] ) #добавляем к слову из переменной words следующее за ним слово 

print(";".join(bigramy))
```

## Задание 2

```python

#импортируем модуль 
import os 

#распечатываем текущую директорию 
print(os.getcwd()) 

#распечатываем список и количество всех файлов в рабочей директории 
for file in os.listdir('.'): 
    print(file) 
print(len(os.listdir('.'))) 

poems = ""
#проходимся циклом по папе poems 
for root, dirs, files in os.walk("poems"): #указываем путь к папке 
    for name in files: 
        with open(os.path.join(root, name), 'r', encoding='utf-8') as f: #открываем файл, указывая путь к файлу, режим доступа к файлу r(только для чтения) и кодировку 
            poems += f.read() + "\n\n" #записываем прочитанный текст в переменную poems 
            
            
with open('poems.txt', 'w', encoding='utf-8') as d: #создаем файл poems, указывалая режим доступа a (для добавления информации в файл) и кодировку 
    d.write(poems) #записываем тексты в файл
```

## Задание 3

```python

stikh = """Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить. 
Скорей, чтобы вылечить кошкину лапу 
Воздушные шарики надо купить! 

И сразу столпился народ на дороге- 
Шумит, и кричит, и на кошку глядит. 
А кошка отчасти идет по дороге, 
Отчасти по воздуху плавно летит!""" 

#количество символов в переменной 
print(len(stikh)) 
```

```python

#очистим текст от пунктуации, используя list compehensions 
words = [word.strip("!?.,-") for word in stikh.split()]
print(words) 

```

```python
#обьединим слова в строку через пробел 
stikh2 = ' '.join(words) 
print(stikh2) 
```

```python
#приведем все слова, что начинались с заглавной буквы, к нижнему регистру 
slova = [] 
for word in stikh2.split(): #проходим циклом по списку, 
    if word.istitle(): 
        slova.append(word.lower()) #заглавные буквы делаем строчными и добавляем в другой список 
    else: 
        slova.append(word) #добавляем остальные слова в список 

print(slova) 
```

```python
#длина текста в словах 
print(len(slova)) 
```

```python
#количество уникальных слов 
ynikalnye = set(slova)
print(len(ynikalnye)) 
```

```python
#последние 10 слов 
print(words[-10:]) 
```

```python
# строки со словами
lines = '\n'.join(words)
print(lines)
```

```python
# биграмы
bigrams = []
# собираем биграммы в список кортежей
for i in range(len(slova)-1):
    bigrams.append((slova[i],slova[i+1]))

# создаем сет уникальных биграм
bi_set = set(bigrams)

# двигаемся по сету
for bigram in bi_set:
    count = 0
    # считаем кол-во повторений уникальной биграммы во всех биграммах
    for j in range(len(bigrams)):
        if (bigram[0]==bigrams[j][0]) and (bigram[1]==bigrams[j][1]):
            count+=1
            
    # выводим биграмму и ее частоту 
    print(bigram," : ",count)
```
